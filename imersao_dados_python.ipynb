```python
import numpy as np

#criação de um dataframe de teste para usar o exemplo de medias.
df_salarios = pd.DataFrame({
    "nome":["Ana","Bruno","Carlos", "Daniele", "Val"],
    "salario": [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a média salarial e substitui os nulos pela média e arredonda os valores
df_salarios["salario_media"]= df_salarios["salario"].fillna(df_salarios["salario"].mean().round(2))

#Calcula a medicana e substitui os nulos pela mediana.
df_salarios["salario_mediana"]= df_salarios["salario"].fillna(df_salarios["salario"].median())

df_salarios

#explicando o código:  df_salarios =BASE / ["salario_media"] = CRIANDO COLUNA SALARIO MEDIA / = df_salarios["salario"] = DENTRO DA BASE IREMOS PREENCHER : OQUE / .fillna(df_salarios["salario"].mean().round(2))  = PREENCHER COM SALARIOS - MEDICA(MEAN), MÉDIA DE SALARIOS. // ROUND = ARREDONDAMENTO, COM DUAS CASAS DECIMAIS.
#median = suavisa os valores medicanos que estão vazios.
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")

renomear_coluna = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df.rename(columns=renomear_coluna, inplace=True);

df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Médio',
    'EN': 'Iniciante',
    'EX': 'Executivo'
})

df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcail',
    'FL': 'Freelancer'
})

df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Hírbido'
})

df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})

df_limpo = df.dropna()
df_limpo = df_limpo.assign(ano=df_limpo["ano"].astype("int64"))
df_limpo.head()
```

```python
import plotly.express as px
```

```python
df_media_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reindex(ordem).reset_index()

fig = px.bar(
    df_media_senioridade,
    x='senioridade',
    y='usd',
    title='Média Salarial por Senioridade',
    labels={'senioridade': 'Senioridade', 'usd': 'Média Salarial Anual em USD'},
    color='senioridade',
    color_discrete_sequence=px.colors.qualitative.Set2
)

fig.update_layout(xaxis_title='Senioridade', yaxis_title='Média Salarial Anual em USD')
fig.show()
```

```python
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']

df_salario_pais_ds = df_data_scientist.groupby('empresa')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(
    df_salario_pais_ds,
    x='empresa',
    y='usd',
    title='Média Salarial para Cientistas de Dados por País',
    labels={'empresa': 'País da Empresa', 'usd': 'Média Salarial Anual em USD'},
    color='empresa',
    color_discrete_sequence=px.colors.qualitative.Plotly
)

fig.update_layout(xaxis_title='País da Empresa', yaxis_title='Média Salarial Anual em USD')
fig.show()
```

```python
# Filtrar apenas para cargos de 'Data Scientist' (já feito no código anterior)
df_data_scientist = df_limpo[df_limpo['cargo'] == 'Data Scientist']

# Calcular a média salarial por país para Cientistas de Dados (já feito)
df_salario_pais_ds = df_data_scientist.groupby('empresa')['usd'].mean().sort_values(ascending=False).reset_index()

# Selecionar o Top 10 países
top_10_paises = df_salario_pais_ds.head(10)

# Calcular a média salarial para os países restantes (fora do Top 10)
media_outros = df_salario_pais_ds.iloc[10:]['usd'].mean()

# Criar um DataFrame para a categoria 'Outros'
df_outros = pd.DataFrame([{'empresa': 'Outros', 'usd': media_outros}])

# Concatenar o Top 10 com a categoria 'Outros'
df_final_plot = pd.concat([top_10_paises, df_outros])

# Criar o gráfico de barras horizontal com Plotly
fig = px.bar(
    df_final_plot,
    y='empresa',  # Eixo Y para os países
    x='usd',      # Eixo X para o salário
    title='Top 10 Países com Maior Salário Médio para Cientistas de Dados',
    labels={'empresa': 'País da Empresa', 'usd': 'Média Salarial Anual em USD'},
    color='empresa',
    color_discrete_sequence=px.colors.qualitative.Plotly,
    orientation='h' # Definir como horizontal
)

fig.update_layout(yaxis_title='País da Empresa', xaxis_title='Média Salarial Anual em USD')
fig.show()
```
